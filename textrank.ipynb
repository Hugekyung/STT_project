```python
from newspaper import Article
article=Article("https://news.naver.com/main/ranking/read.nhn?mid=etc&sid1=111&rankingType=popular_day&oid=296&aid=0000046738&date=20201011&type=1&rankingSeq=2&rankingSectionId=103", language='ko')
article.download()
article.parse()
article.text
```

```python
from konlpy.tag import Kkma
kkma=Kkma()
sen=kkma.sentences(article.text)
```

```python
from textrank.textrank import KeysentenceSummarizer
summerizer=KeysentenceSummarizer(tokenize=kkma.morphs)
```

```python
sen
```

```python
list(zip(*summerizer.summarize(sents=sen,topk=3)))[2]
```

```python
from gensim.summarization.summarizer import summarize
text="\n".join(sen)
summarize(text,ratio=0.09).split("\n")
```

```python
from krwordrank.sentence import summarize_with_sentences
summarize_with_sentences(sen,num_keysents=3)[1]
```
